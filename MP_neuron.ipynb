# MP Neuron class

```python
class MP_neuron:

  def __init__(self):
    self.threshold=threshold

#1. model
  def model(self,x):
    ## the model is simple MP neuron.returns 1, if the sum of all inputs is greater than the threshold values, else zero
    ## here the activation function is a step function, based on the threshold values
    return(np.sum(x)>=self.threshold)

#2. prediction of the output using the model
## using the model predicst the output for a given set of inputs
  def predict(self,X):
    Y=[] # creating a empty list to store all the predicted Y uisng the MP neuron model
    for x in X: #X should be the input data set , could be either train data or test data
        result=self.model(x) ## the model function is called to evaluate the input data row wise. the input data should be binarised as it is an MP neuron
        Y.append(result)
    return np.array(Y) ## the ouptput contains zero or ones

#3.Learning algortihm for parameter optimisation, here it is about optimising the parameter 'b'. in the case of a MP neuron the weight,'w' is one
#so this need not be optimised
  def fit(self,X,Y):
    accuracy={}
    # the loop below finds the accuracy of the model for all the possible threshold
    #the rows of the data matrix contains 0, and 1. so if we choose a row and do a coloumnwise summation, the maximum possible value of that would be equal to number of coloumn
    #Hence in the below loop differnt threshold values are tried in the MP neuron model and accuracy corresponding to each value is computed

    for threshold in range (X.shape[1]+1):
      self.threshold=threshold
      Y_pred=self.predict(X)
      accuracy[threshold]=accuracy_score(Y_pred,Y) # assiginnig value to each key
    # form the accuracy calculated for all the threshold, the value of thrshold corresponding to maximum accuracy is fetched
    best_threshold=max(accuracy,key=accuracy.get)
    # storing the value of the threshold to the variable
    self.threshold=best_threshold
    print('optimal value of b is', best_b)
    print('Highest accuracy is ', accuracy[best_b])


```
